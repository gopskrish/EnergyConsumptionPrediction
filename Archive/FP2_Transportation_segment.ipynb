```python
p = [0,1,2,3,4,5]
q = [0,1,2,3,5]
d = [0,1]
pdq=[]
seasonal_pdq=[]
for p_ in p:
  for q_ in q:
    for d_ in d:
      pdq.append((p_,d_,q_))
      seasonal_pdq.append((p_,d_,q_))
```

```python
import warnings
warnings.filterwarnings("ignore")
import tqdm.notebook as tqdm
import itertools
from sklearn import metrics

def arimaModel(data, params, seasonal_param):
  return ARIMA(data, order = params)

def sarimaxModel(data, params, seasonal_param):
  return SARIMAX(endog=data,
                order =params,
                seasonal_order = seasonal_param,
                mle_regression=True,
                simple_differencing=True
               )

def HyperparameersForModels(data, modelfunction):

  results_list=[]

  for param in tqdm.tqdm(pdq):
      for seasonal_param in seasonal_pdq:
          fin_res={}
          try:

              model = modelfunction(data, param, seasonal_param)
              results = model.fit()

              if(modelfunction.startswith('arima')):
                fin_res['ARIMA']='({})'.format(param)
              else:
                fin_res['SARIMAX']='({}),({})'.format(param,seasonal_param)


              val_srt = len(data)-len(results.forecasts[0])

              fin_res['MAPE']=metrics.mean_absolute_percentage_error(data.values[val_srt:],(results.forecasts.reshape(len(results.forecasts[0]),)))
              fin_res['MAE']=metrics.mean_absolute_error(data.values[val_srt:],(results.forecasts.reshape(len(results.forecasts[0]),)))
              fin_res['AIC']=pd.read_html(((results.summary().tables[0])).as_html())[0].iloc[2,3]
              fin_res['BIC']=pd.read_html(((results.summary().tables[0])).as_html())[0].iloc[3,3]
              results_list.append(fin_res)

          except:
              pass
  return results_list

def display_result(result):
  fin_results = pd.DataFrame(result).set_index("ARIMA")
  fin_results = fin_results[~fin_results.index.duplicated(keep='first')]
  fin_results_rank = fin_results.rank(method = 'dense')#.sum(axis=1)
  fin_results_rank['COMBINED_RANK (Sum of Ranks, Lower is Better)']=fin_results_rank.sum(axis=1)

  fin_results.merge(fin_results_rank,left_index=True,right_index=True,suffixes=( ' Value', ' Column Rank')).sort_values('COMBINED_RANK (Sum of Ranks, Lower is Better)',ascending=True).head(20).style.format('{:,.1f}').background_gradient(axis=0,
    cmap='YlOrRd',subset = ['COMBINED_RANK (Sum of Ranks, Lower is Better)'])
  print(fin_results)
```
