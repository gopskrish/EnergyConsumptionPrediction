#Data Preprocessing

```python
import pandas as pd
df2015_2023 = pd.read_csv('/content/sample_data/Group_11_Short Term EDA.csv')
```

```python
median_df = df2015_2023.median(numeric_only = 'True')
#print(median_df)
print(df2015_2023)
```

```python
#Removing first two rows
df2015_2023 = df2015_2023.drop(labels=[0,1], axis=0)
#df2021_2023.columns('Year','Month','State','Data Status','Residential Revenue','Residential Sales','Residential Customers','Residential Price','Commercial Revenue','Commercial Sales','Commercial Customers','Commercial Price','Industrial Revenue','Industrial Sales','Industrial Customers','Industrial Price','Transportation Revenue','Transportation Sales','Transportation Customers','Transportation Price','Total Revenue','Total Sales','Total Customers','Total Price')
```

```python
#Naming Columns
df2015_2023.rename(columns = {'Unnamed: 0':'Year','Unnamed: 1':'Month','Unnamed: 2':'State','Unnamed: 3':'Data Status','RESIDENTIAL':'Residential Revenue','Unnamed: 5':'Residential Sales','Unnamed: 6':'Residential Customers','Unnamed: 7':'Residential Price','COMMERCIAL':'Commercial Revenue','Unnamed: 9':'Commercial Sales','Unnamed: 10':'Commercial Customers','Unnamed: 11':'Commercial Price','INDUSTRIAL':'Industrial Revenue','Unnamed: 13':'Industrial Sales','Unnamed: 14':'Industrial Customers','Unnamed: 15':'Industrial Price','TRANSPORTATION':'Transportation Revenue','Unnamed: 17':'Transportation Sales','Unnamed: 18':'Transportation Customers','Unnamed: 19':'Transportation Price','TOTAL':'Total Revenue','Unnamed: 21':'Total Sales','Unnamed: 22':'Total Customers','Unnamed: 23':'Total Price'}, inplace = True)
print(df2015_2023)
```

```python
#Changing str to int
df2015_2023['Residential Revenue'] = df2015_2023['Residential Revenue'].str.replace(',', '').astype(int)
```

```python
#Changing str to int
df2015_2023['Residential Sales'] = df2015_2023['Residential Sales'].str.replace(',', '').astype(int)
df2015_2023['Residential Customers'] = df2015_2023['Residential Customers'].str.replace(',', '').astype(int)
df2015_2023['Commercial Revenue'] = df2015_2023['Commercial Revenue'].str.replace(',', '').astype(int)
df2015_2023['Commercial Sales'] = df2015_2023['Commercial Sales'].str.replace(',', '').astype(int)
df2015_2023['Commercial Customers'] = df2015_2023['Commercial Customers'].str.replace(',', '').astype(int)
df2015_2023['Industrial Revenue'] = df2015_2023['Industrial Revenue'].str.replace(',', '').astype(int)
df2015_2023['Industrial Sales'] = df2015_2023['Industrial Sales'].str.replace(',', '').astype(int)
df2015_2023['Industrial Customers'] = df2015_2023['Industrial Customers'].str.replace(',', '').astype(int)
df2015_2023['Total Revenue'] = df2015_2023['Total Revenue'].str.replace(',', '').astype(int)
df2015_2023['Total Sales'] = df2015_2023['Total Sales'].str.replace(',', '').astype(int)
df2015_2023['Total Customers'] = df2015_2023['Total Customers'].str.replace(',', '').astype(int)
print(df2015_2023)
```

```python
datatypes = df2015_2023.dtypes
print(datatypes)
```

```python
#Changing object datatypes 
df2015_2023['Total Price'] = df2015_2023['Total Price'].astype(float)
df2015_2023['Residential Price'] = df2015_2023['Residential Price'].astype(float)
df2015_2023['Commercial Price'] = df2015_2023['Commercial Price'].astype(float)
df2015_2023['Industrial Price'] = df2015_2023['Industrial Price'].astype(float)
df2015_2023['Transportation Revenue'] = df2015_2023['Transportation Revenue'].astype(int)
#df2015_2023['Transportation Sales'] = df2015_2023['Transportation Sales'].astype(int)
df2015_2023['Transportation Customers'] = df2015_2023['Transportation Customers'].astype(int)
df2015_2023['Transportation Price'] = df2015_2023['Transportation Price'].astype(float)
datatypes = df2015_2023.dtypes
print(datatypes)
print(df2015_2023)
```

```python
df2015_2023['Transportation Sales'] = df2015_2023['Transportation Sales'].str.replace(',', '').astype(int)
df = df2015_2023
```

```python
#df['Year_Month']=df['Year']*100+df['Month']
#df['Year_Month']=df['Year_Month'].astype(str)
```

```python
df['Year_Month1'] = pd.to_datetime(df.Year.astype(str) + '-' + df.Month.astype(str))
print(df)
```

```python
#Time Series Graph for total sales
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from IPython.core.display import display, HTML,Markdown
display(HTML("<style>.container { width:100% !important; }</style>"))


ax=df.groupby('Year_Month1')['Total Sales'].sum().plot(figsize=(16,6),marker='o',grid=True)
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0f'))
plt.title('Time Series Graph of Total Monthly Sales')
```
